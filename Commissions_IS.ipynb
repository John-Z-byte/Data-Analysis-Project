```python
import pandas as pd

# Load the CSV files into pandas DataFrames
df_atd = pd.read_csv('Team_1_John_2025(ATD).csv', encoding='ISO-8859-1')
df_conv = pd.read_csv('Team_1_John_2025(CONV).csv', encoding='ISO-8859-1')
df_sales = pd.read_csv('Team_1_John_2025(SALES).csv', encoding='ISO-8859-1')

# Check the first few rows of each DataFrame to ensure they loaded correctly
print("ATD Data:")
print(df_atd.head())

print("CONV Data:")
print(df_conv.head())

print("SALES Data:")
print(df_sales.head())
```

```python
#Cleaning proccess

# Drop unnamed columns (columns that start with 'Unnamed:')
df_sales_cleaned = df_sales.loc[:, ~df_sales.columns.str.contains('^Unnamed')]
df_atd_cleaned = df_atd.loc[:, ~df_atd.columns.str.contains('^Unnamed')]
df_conv_cleaned = df_conv.loc[:, ~df_conv.columns.str.contains('^Unnamed')]

# Print the cleaned data
print("Cleaned ATD Data:")
print(df_atd_cleaned.head())

print("Cleaned CONV Data:")
print(df_conv_cleaned.head())

print("Cleaned SALES Data:")
print(df_sales_cleaned.head())
```

```python
#ROW COMISSIONS
# Step 1: Convert the 'DATE' column to datetime format (if not already done)
df_sales_cleaned['DATE'] = pd.to_datetime(df_sales_cleaned['DATE'], errors='coerce')

# Step 2: Define the start and end date for the week (6th January - 10th January)
start_date = pd.to_datetime('2025-01-06')
end_date = pd.to_datetime('2025-01-10')

# Step 3: Filter the sales data for the specific week (6th January - 10th January)
filtered_sales = df_sales_cleaned[(df_sales_cleaned['DATE'] >= start_date) & (df_sales_cleaned['DATE'] <= end_date)]

# Step 4: Ensure the 'COMISSION' column is in numeric format
filtered_sales['COMISSION'] = filtered_sales['COMISSION'].replace({r'[^0-9.]': ''}, regex=True).astype(float)

# Step 5: Sum the commissions for the filtered data
total_commissions_week = filtered_sales['COMISSION'].sum()

# Step 6: Optionally, you can also calculate commissions by rep
commissions_by_rep_week = filtered_sales.groupby('REP')['COMISSION'].sum().reset_index()

# Print the total commissions for the week
print(f"Total commissions from {start_date.date()} to {end_date.date()}: ${total_commissions_week:.2f}")

# Print the commissions by rep for the specific week
print("Commissions by rep for the week:")
print(commissions_by_rep_week)
```

```python
# Strip any leading or trailing spaces in column names
df_sales_cleaned.columns = df_sales_cleaned.columns.str.strip()
df_atd_cleaned.columns = df_atd_cleaned.columns.str.strip()
df_conv_cleaned.columns = df_conv_cleaned.columns.str.strip()

# Convert 'DATE ADDED' columns to datetime and handle errors
df_sales_cleaned['DATE'] = pd.to_datetime(df_sales_cleaned['DATE'], errors='coerce')
df_atd_cleaned['DATE ADDED'] = pd.to_datetime(df_atd_cleaned['DATE ADDED'], errors='coerce')
df_conv_cleaned['DATE ADDED'] = pd.to_datetime(df_conv_cleaned['DATE ADDED'], errors='coerce')

# Verify the data types of the columns to ensure they are datetime64[ns]
print(df_sales_cleaned['DATE'].dtype)  # Should output datetime64[ns]
print(df_atd_cleaned['DATE ADDED'].dtype)  # Should output datetime64[ns]
print(df_conv_cleaned['DATE ADDED'].dtype)  # Should output datetime64[ns]

# Use .loc to avoid SettingWithCopyWarning when modifying columns
df_sales_cleaned.loc[:, 'DATE'] = pd.to_datetime(df_sales_cleaned['DATE'], errors='coerce')
df_atd_cleaned.loc[:, 'DATE ADDED'] = pd.to_datetime(df_atd_cleaned['DATE ADDED'], errors='coerce')
df_conv_cleaned.loc[:, 'DATE ADDED'] = pd.to_datetime(df_conv_cleaned['DATE ADDED'], errors='coerce')

```

```python
# First, we will convert the 'DATE' column to datetime if it's not already done
df_sales_cleaned['DATE'] = pd.to_datetime(df_sales_cleaned['DATE'], errors='coerce')

# Set the start and end dates for the week you are interested in (example: 6 Jan - 10 Jan)
start_date = pd.to_datetime('2025-01-06')
end_date = pd.to_datetime('2025-01-10')

# Filter sales data for the specified week
weekly_sales_data = df_sales_cleaned[(df_sales_cleaned['DATE'] >= start_date) & (df_sales_cleaned['DATE'] <= end_date)]

# Group by Rep and calculate total commission for each rep within that week
weekly_sales_commission = weekly_sales_data.groupby('REP')['COMISSION'].sum().reset_index()

# Print weekly commission data
print(weekly_sales_commission)
```

```python
# Convert 'DATE ADDED' to datetime if not already done
df_conv_cleaned['DATE ADDED'] = pd.to_datetime(df_conv_cleaned['DATE ADDED'], errors='coerce')

# Filter conversion data for the same week (6 Jan - 10 Jan)
weekly_conv_data = df_conv_cleaned[(df_conv_cleaned['DATE ADDED'] >= start_date) & (df_conv_cleaned['DATE ADDED'] <= end_date)]

# Group by Rep and calculate SUM of SALES and SUM of ATD
weekly_conversion_data = weekly_conv_data.groupby('REP').agg(
    total_sales=('SALES', 'sum'),
    total_atd=('ATD', 'sum')
).reset_index()

# Calculate the conversion percentage
weekly_conversion_data['CONVERSION'] = (weekly_conversion_data['total_sales'] / weekly_conversion_data['total_atd']) * 100

# Print weekly conversion data
print(weekly_conversion_data)
```

```python
# Commission calculation function based on conversion percentage
def calculate_commission_percentage(conversion):
    # Round the conversion value to handle any small floating-point discrepancies
    conversion = round(conversion)

    if conversion >= 85:
        return 1.25  # 125%
    elif conversion == 84:
        return 1.20  # 120%
    elif conversion == 83:
        return 1.15  # 115%
    elif conversion == 82:
        return 1.10  # 110%
    elif conversion == 81:
        return 1.05  # 105%
    elif conversion == 80:
        return 1.00  # 100%
    elif conversion == 79:
        return 0.95  # 95%
    elif conversion == 78:
        return 0.90  # 90%
    elif conversion == 77:
        return 0.85  # 85%
    elif conversion == 76:
        return 0.80  # 80%
    elif conversion == 75:
        return 0.75  # 75%
    elif conversion == 74:
        return 0.70  # 70%
    elif conversion == 73:
        return 0.65  # 65%
    elif conversion == 72:
        return 0.60  # 60%
    elif conversion == 71:
        return 0.55  # 55%
    elif conversion == 70:
        return 0.50  # 50%
    else:
        return 0.00  # No commission

# Apply the commission percentage calculation again
weekly_conversion_data['commission_percentage'] = weekly_conversion_data['CONVERSION'].apply(calculate_commission_percentage)

# Check the result
print(weekly_conversion_data[['REP', 'total_sales', 'total_atd', 'CONVERSION', 'commission_percentage']])
```

```python
# Step 1: Group the sales data by rep and calculate the total commission
weekly_sales_commission = df_sales_cleaned[(df_sales_cleaned['DATE'] >= start_date) & (df_sales_cleaned['DATE'] <= end_date)].groupby('REP')['COMISSION'].sum().reset_index()

# Step 2: Merge the weekly sales commission with weekly conversion data based on 'REP'
df_commissions = pd.merge(weekly_sales_commission, weekly_conversion_data, how='inner', on='REP')

# Step 3: Define a function to apply the commission percentage based on conversion
def calculate_commission_percentage(conversion):
    if conversion >= 85:
        return 1.25  # 125%
    elif conversion == 84:
        return 1.20  # 120%
    elif conversion == 83:
        return 1.15  # 115%
    elif conversion == 82:
        return 1.10  # 110%
    elif conversion == 81:
        return 1.05  # 105%
    elif conversion == 80:
        return 1.00  # 100%
    elif conversion == 79:
        return 0.95  # 95%
    elif conversion == 78:
        return 0.90  # 90%
    elif conversion == 77:
        return 0.85  # 85%
    elif conversion == 76:
        return 0.80  # 80%
    elif conversion == 75:
        return 0.75  # 75%
    elif conversion == 74:
        return 0.70  # 70%
    elif conversion == 73:
        return 0.65  # 65%
    elif conversion == 72:
        return 0.60  # 60%
    elif conversion == 71:
        return 0.55  # 55%
    elif conversion == 70:
        return 0.50  # 50%
    else:
        return 0.00  # No commission

# Step 4: Apply the commission percentage to the commission data
df_commissions['FINAL_COMMISSION'] = df_commissions['COMISSION'] * df_commissions['commission_percentage']

# Step 5: Print the final result
print(df_commissions[['REP', 'COMISSION', 'commission_percentage', 'FINAL_COMMISSION']])
```
